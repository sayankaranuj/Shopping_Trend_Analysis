```python
import pandas as pd

# Load the data file (make sure the file is in the same folder as your notebook)
df = pd.read_csv('shopping_trends.csv')

# Display the first few rows of the data
df.head()
```

```python
# Check the number of rows and columns
print("Dataset shape:", df.shape)

# Display the column names
print("Column names:", df.columns.tolist())

# Look for missing values
print("\nMissing values:\n", df.isnull().sum())
```

```python
# Show summary statistics for numerical columns
print("Summary statistics:\n", df.describe())

# Show the first few rows of the dataset again for reference
print("\nSample data:\n", df.head())
```

```python
import matplotlib.pyplot as plt

# Group by category and calculate the total purchase amount
top_categories = df.groupby('Category')['Purchase Amount (USD)'].sum().sort_values(ascending=False)

# Plot a bar chart for the top-selling categories
top_categories.plot(kind='bar', figsize=(8, 6), color='skyblue')
plt.title('Top-Selling Categories')
plt.xlabel('Category')
plt.ylabel('Total Purchase Amount (USD)')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Count the occurrences of each payment method
payment_counts = df['Payment Method'].value_counts()

# Plot a pie chart for payment methods
payment_counts.plot(kind='pie', figsize=(8, 8), autopct='%1.1f%%', colors=['#ff9999', '#66b3ff', '#99ff99', '#ffcc99'])
plt.title('Distribution of Payment Methods')
plt.ylabel('')  # Remove the y-axis label for a cleaner look
plt.show()
```

```python
# Create age groups
bins = [0, 20, 30, 40, 50, 60, 70, 100]
labels = ['<20', '20-29', '30-39', '40-49', '50-59', '60-69', '70+']
df['Age Group'] = pd.cut(df['Age'], bins=bins, labels=labels, right=False)

# Calculate average purchase amount by age group
avg_purchase_by_age = df.groupby('Age Group')['Purchase Amount (USD)'].mean()

# Plot a bar chart for average purchase by age group
plt.figure(figsize=(8, 6))
avg_purchase_by_age.plot(kind='bar', color='orange')
plt.title('Average Purchase Amount by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Average Purchase Amount (USD)')
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Add a caption explaining the chart
plt.figtext(0.5, -0.1, 'This chart shows the average purchase amount for each age group.', ha='center', fontsize=12)

# Display the plot
plt.show()
```

```python
# Calculate total purchase amount by category
purchase_by_category = df.groupby('Category')['Purchase Amount (USD)'].sum()

# Plot a bar chart for purchase amount by category
plt.figure(figsize=(8, 6))
purchase_by_category.plot(kind='bar', color='green')
plt.title('Total Purchase Amount by Product Category')
plt.xlabel('Product Category')
plt.ylabel('Total Purchase Amount (USD)')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Add a caption explaining the chart
plt.figtext(0.5, -0.1, 'This chart shows the total purchase amount for each product category.', ha='center', fontsize=12)

# Display the plot
plt.show()
```

```python
# Calculate the average review rating by category
avg_rating_by_category = df.groupby('Category')['Review Rating'].mean()

# Plot a bar chart for average review rating by category
plt.figure(figsize=(8, 6))
avg_rating_by_category.plot(kind='bar', color='purple')
plt.title('Average Review Rating by Product Category')
plt.xlabel('Product Category')
plt.ylabel('Average Review Rating')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Add a caption explaining the chart
plt.figtext(0.5, -0.1, 'This chart shows the average review rating for each product category.', ha='center', fontsize=12)

# Display the plot
plt.show()
```

```python
# Scatter plot showing the relationship between previous purchases and current purchase amount
plt.figure(figsize=(8, 6))
plt.scatter(df['Previous Purchases'], df['Purchase Amount (USD)'], color='teal', alpha=0.5)
plt.title('Correlation Between Previous Purchases and Current Purchase Amount')
plt.xlabel('Previous Purchases')
plt.ylabel('Current Purchase Amount (USD)')
plt.grid(True)

# Add a caption explaining the chart
plt.figtext(0.5, -0.1, 'This scatter plot shows the correlation between previous purchases and current spending.', ha='center', fontsize=12)

# Display the plot
plt.show()
```

```python
# Count the occurrences of each shipping type
shipping_counts = df['Shipping Type'].value_counts()

# Plot a pie chart for shipping types
plt.figure(figsize=(8, 8))
shipping_counts.plot(kind='pie', autopct='%1.1f%%', colors=['#ffcc99', '#66b3ff', '#ff9999', '#99ff99'])
plt.title('Distribution of Shipping Types')
plt.ylabel('')  # Remove the y-axis label for a cleaner look

# Add a caption explaining the chart
plt.figtext(0.5, -0.1, 'This pie chart shows the distribution of shipping types selected by customers.', ha='center', fontsize=12)

# Display the plot
plt.show()
```

```python
# Count the occurrences of each subscription status
subscription_counts = df['Subscription Status'].value_counts()

# Plot a pie chart for subscription status
plt.figure(figsize=(8, 8))
subscription_counts.plot(kind='pie', autopct='%1.1f%%', colors=['#ffcc00', '#66cc66'])
plt.title('Breakdown of Customers by Subscription Status')
plt.ylabel('')  # Remove the y-axis label for a cleaner look

# Add a caption explaining the chart
plt.figtext(0.5, -0.1, 'This pie chart shows the breakdown of customers by their subscription status.', ha='center', fontsize=12)

# Display the plot
plt.show()
```

```python
# Calculate the average purchase amount for customers with and without a discount
purchase_with_discount = df[df['Discount Applied'] == 'Yes']['Purchase Amount (USD)']
purchase_without_discount = df[df['Discount Applied'] == 'No']['Purchase Amount (USD)']

# Plot a box plot to compare purchase amounts with and without discount
plt.figure(figsize=(8, 6))
plt.boxplot([purchase_with_discount, purchase_without_discount], labels=['Discount Applied', 'No Discount'], patch_artist=True, boxprops=dict(facecolor='lightblue', color='blue'))
plt.title('Effect of Discount on Purchase Amount')
plt.ylabel('Purchase Amount (USD)')
plt.grid(True)

# Add a caption explaining the chart
plt.figtext(0.5, -0.1, 'This box plot compares the purchase amounts of customers who applied a discount versus those who did not.', ha='center', fontsize=12)

# Display the plot
plt.show()
```

```python
# Create a cross-tab of payment method and frequency of purchases
payment_method_vs_frequency = pd.crosstab(df['Preferred Payment Method'], df['Frequency of Purchases'])

# Plot a bar chart for payment method vs frequency of purchases
payment_method_vs_frequency.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.title('Payment Method vs Frequency of Purchases')
plt.xlabel('Payment Method')
plt.ylabel('Number of Customers')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Add a caption explaining the chart
plt.figtext(0.5, -0.1, 'This stacked bar chart shows how payment methods vary with frequency of purchases.', ha='center', fontsize=12)

# Display the plot
plt.show()
```
